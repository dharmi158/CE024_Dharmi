```python
from sklearn import preprocessing
from sklearn.naive_bayes import GaussianNB, MultinomialNB
```

```python
import numpy as np
import pandas as pd
```

```python
data1 = pd.read_csv('Dataset1.csv')
#print(data)
outlook = data1['Outlook'].values
temp = data1['Temp'].values
humidity = data1['Humidity'].values
wind = data1['Wind'].values
play = data1['Play'].values
```

```python
le = preprocessing.LabelEncoder()

outlook_encoded = le.fit_transform(outlook)
temp_encoded = le.fit_transform(temp)
humidity_encoded = le.fit_transform(humidity)
wind_encoded = le.fit_transform(wind)
label = le.fit_transform(play)

print("Outlook Encoded :" , outlook_encoded)
print("Temp Encoded :" , temp_encoded)
print("Humidity Encoded :" , humidity_encoded)
print("Wind Encoded :" , wind_encoded)
print("Play :" , label)
```

```python
f1=tuple(zip(outlook_encoded,temp_encoded , humidity_encoded , wind_encoded))
print("Features:",f1)
```

```python
model = MultinomialNB()
model.fit(f1,label)
```

```python
predicted= model.predict([[0,1,2,0]]) # Rainy , mild , High , F 
print("Predicted Value:", predicted)
```

CHECK FOR ACCURACY

```python
from sklearn.model_selection import train_test_split
```

```python
data_train, data_test, target_train, target_test = train_test_split(f1,label, test_size = 0.20, random_state = 23)
```

Accuracy FOR gaussianNB model

```python
#Accuracy FOR gaussianNB model
model = MultinomialNB()
model.fit(f1 , label)
target_pred_mo = model.predict(data_test)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(target_test, target_pred_mo))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(target_test, target_pred_mo)
```

```python
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score

precision = precision_score(target_test, target_pred_mo, average=None)
recall = recall_score(target_test, target_pred_mo, average=None)

print('precision: {}'.format(precision))
print('recall: {}'.format(recall))
```
